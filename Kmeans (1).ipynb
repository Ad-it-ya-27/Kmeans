```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/Student dataset working 1.csv')
print(df.head())
```

Here, I am representing gender as 0-female 1-male
In the end the result column shows the happiness score of the children with respect to their performance in the examination.

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
label = {'female': 0, 'male': 1}

print(x)
print(y)
```

```python
#here we are clusterint the result(happiness score of the student with respect to the social score into 3 clusters)
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
plt.xlabel("Social Score")
plt.ylabel("Happiness Score")
plt.scatter(x.iloc[:,-1], y, c=kmeans.labels_)
plt.show()
```

```python
#here we are clusterint the result(happiness score of the student with respect to the social score into 4 clusters)
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
plt.xlabel("Social Score")
plt.ylabel("Happiness Score")
plt.scatter(x.iloc[:,-1], y, c=kmeans.labels_)
plt.show()
```

Output Analysis-

After seeing the output of the 3 cluster and 4 cluster we can see that objects that are plotted together are grouped into a single cluster denoted by colour

Moreover we can see that the students who have just cleared the paper(by scoring marks very near to 40)show the highest happiness score

When the same data is plotted into 4 clusters from 3 we see only one point in the new cluster this is because of non uniform distribution of data in the dataset(due to an outlier).
